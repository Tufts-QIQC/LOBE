## Yukawa Interaction 

\begin{align}
g\bar\psi \psi \phi&= \frac{gm}{\sqrt{2L}}\sum_{k_1, k_2, k_3} (\delta_{k_3 + k_2, k_1}\delta_{s_1, s_2}\sqrt{k_1 k_2}\left(\frac{1}{k_1} + \frac{1}{k_2} \right)b_1^\dagger b_2 a_3 \\ \nonumber
& + \delta_{k_2, k_3 + k_1}\delta_{s_1, s_2}\sqrt{k_1 k_2}\left(\frac{1}{k_1} + \frac{1}{k_2} \right)b_1^\dagger b_2 a_3^\dagger\\ \nonumber
&+\delta_{k_3, k_2 + k_1}\delta_{s_1, -s_2}\sqrt{k_1 k_2}\left(-\frac{1}{k_2} + \frac{1}{k_1} \right)b_1^\dagger d_2^\dagger a_3\\ \nonumber
&-\delta_{k_3 + k_2, k_1}\delta_{s_1, s_2}\sqrt{k_1 k_2}\left(\frac{1}{k_1} + \frac{1}{k_2} \right)d_2^\dagger d_1 a_3^\dagger\\ \nonumber
&- \delta_{k_3 + k_1, k_2}\delta_{s_1, s_2}\sqrt{k_1 k_2}\left(\frac{1}{k_1} + \frac{1}{k_2} \right)d_2^\dagger d_1 a_3\\ \nonumber
&-\delta_{k_1 + k_2, k_3}\delta_{s_1, -s_2}\sqrt{k_1 k_2}\left(-\frac{1}{k_2} + \frac{1}{k_1} \right)b_2 d_1 a_3^\dagger

)
\end{align}

g^2\bar\psi\phi\frac{\gamma^+}{i\partial^+}\phi\psi &= \frac{g^2}{2L}\sum_{k_1, k_2, k_3, k_4}\\

```python
def H_free(I, mf = 1, mb = 1, L = 10):
    hamiltonian_free = ParticleOperator({})

    for i, k in enumerate(np.arange(1/2, I, 1)):
        bstr = 'b' + str(i) + "^ " + "b" + str(i)
        dstr = 'd' + str(i) + "^ " + "d" + str(i)
        hamiltonian_free += (mf**2 / k) * ParticleOperator(bstr) +\
                            (mf**2 / k) * ParticleOperator(dstr)
    for i, k, in enumerate(np.arange(1, I, 1)):
        astr = 'a' + str(i) + "^ " + "a" + str(i)
        hamiltonian_free += (mb**2 / k) * ParticleOperator(astr) 

    return 1/np.pi * hamiltonian_free
```

```python
def three_point(I, m = 1, g = 1, L = 10):
    Y_int = ParticleOperator({})

    for s1 in [1, -1]:
        for s2 in [1, -1]:
            for i1, k1 in enumerate(np.arange(1/2, I, 1)):
                for i2, k2 in enumerate(np.arange(1/2, I, 1)):
                    k1k2prod = np.sqrt(k1 * k2) 
                    for i3, k3 in enumerate(np.arange(1, I, 1)):
                        Y_int += k1k2prod * (1/k1 + 1/k2) * int(k3 + k2 == k1) * int(s1 == s2)* ParticleOperator('b' + str(i1) + "^ " +
                                                                             'b' + str(i2) + 
                                                                             " a" + str(i3)) 
                        Y_int += k1k2prod * (1/k1 + 1/k2) * int(k1 + k3 == k2) * int(s1 == s2)* ParticleOperator('b' + str(i1) + "^ " +
                                                                             'b' + str(i2) + 
                                                                             " a" + str(i3) + '^')
                        Y_int += k1k2prod * (1/k1 - 1/k2) * int(k1 + k2 == k3) * int(s1 == -s2)* ParticleOperator('b' + str(i1) + "^ " +
                                                                             'd' + str(i2) + "^ "
                                                                             "a" + str(i3))
                        Y_int -= k1k2prod * (1/k1 + 1/k2) * int(k3 + k2 == k1) * int(s1 == s2)* ParticleOperator('d' + str(i2) + "^ " +
                                                                             'd' + str(i1) + "^ "
                                                                             "a" + str(i3) + "^")
                        Y_int -= k1k2prod * (1/k1 + 1/k2) * int(k3 + k1 == k2) * int(s1 == s2)* ParticleOperator('d' + str(i2) + "^ " +
                                                                             'd' + str(i1) + 
                                                                             " a" + str(i3))
                        Y_int -= k1k2prod * (1/k1 - 1/k2) * int(k1 + k2 == k3) * int(s1 == s2)* ParticleOperator('b' + str(i2) +
                                                                             ' d' + str(i1) + 
                                                                             " a" + str(i3) + "^")
                    
    return g * m / np.sqrt(2 * L) * Y_int
```

```python
def four_point(I, m = 1, g = 1, L = 10):
    Y_four = ParticleOperator({})

    for s1 in [1, -1]:
        for s2 in [1, -1]:
            for i1, k1 in enumerate(np.arange(1/2, I, 1)):
                for i2, k2 in enumerate(np.arange(1/2, I, 1)):
                    for i3, k3 in enumerate(np.arange(1, I, 1)):
                        for i4, k4 in enumerate(np.arange(1, I, 1)):
                            
                            Y_four += np.sqrt(k1 * k4) / (k3 + k4) * (k1 + k2 == k4 == k3) * (s1 == s2) * ParticleOperator(
                                'b' + str(i1) + '^ a' + str(i2) + '^ a' + str(i3) + ' b' + str(i4)
                            )
                            Y_four += np.sqrt(k1 * k4) / (k3 + k4) * (k3 + k4 + k2 == k1) * (s1 == -s2) * ParticleOperator(
                                'b' + str(i1) + '^ a' + str(i2) + ' a' + str(i3) + ' b' + str(i4)
                            )
                            Y_four += np.sqrt(k1 * k4) / (k3 + k4) * (k2 == k3 + k1 + k4) * (s1 == -s2) * ParticleOperator(
                                'a' + str(i2) + '^ a' + str(i3) + ' d' + str(i1) + ' b' + str(i4)
                            )
                            Y_four += np.sqrt(k1 * k4) / (k3 + k4) * (k1 + k4 + k3 == k2) * (s1 == -s2) * ParticleOperator(
                                'b' + str(i1) + '^ d' + str(i4) + ' a' + str(i3) + '^ a' + str(i2)
                            )
                            Y_four -= np.sqrt(k1 * k4) / (k3 + k4) * (k4 + k3 + k2 == k1) * (s1 == s2) * ParticleOperator(
                                'd' + str(i4) + '^ a' + str(i2) + '^ a' + str(i3) + '^ d' + str(i1)
                            )
                            Y_four -= np.sqrt(k1 * k4) / (k3 + k4) * (k4 + k3 == k2 + k1) * (s1 == s2) * ParticleOperator(
                                'd' + str(i4) + '^ a' + str(i3) + '^ a' + str(i2) + ' d' + str(i1)
                            )
                            # Y_four += np.sqrt(k1 * k4) / (-k3 + k4) * (k1 + k3 == k4 + k2) * (s1 == s2) * ParticleOperator(
                            #     'b' + str(i1) + '^ a' + str(i3) + '^ a' + str(i2) + ' b' + str(i4)
                            # )
                            # Y_four += np.sqrt(k1 * k4) / (-k3 + k4) * (k2 + k3 == k4 + k1) * (s1 == -s2) * ParticleOperator(
                            #     'a' + str(i2) + '^ a' + str(i3) + '^ d' + str(i1) + ' b' + str(i4)
                            # )
                            # Y_four += np.sqrt(k1 * k4) / (-k3 + k4) * (k1 + k2 + k3 == k4) * (s1 == s2) * ParticleOperator(
                            #     'b' + str(i1) + '^ a' + str(i2) + '^ a' + str(i3) + '^ b' + str(i2)
                            # )
                            # Y_four += np.sqrt(k1 * k4) / (-k3 + k4) * (k1 + k4 + k3 == k2) * (s1 == -s2) * ParticleOperator(
                            #     'b' + str(i1) + '^ d' + str(i4) + '^ a' + str(i3) + '^ d' + str(i1)
                            # )
                            # Y_four += np.sqrt(k1 * k4) / (k3 - k4) * (k1 + k4 + k2 == k3) * (s1 == -s2) * ParticleOperator(
                            #     'b' + str(i1) + '^ d' + str(i4) + '^ a' + str(i2) + ' a' + str(i3)
                            # )
                            # Y_four += np.sqrt(k1 * k4) / (k3 - k4) * (k1 + k4 + k2 == k3) * (s1 == -s2) * ParticleOperator(
                            #     'b' + str(i1) + '^ d' + str(i4) + '^ a' + str(i2) + '^ a' + str(i3)
                            # )
                            # Y_four += np.sqrt(k1 * k4) / (k3 - k4) * (k1 + k4 + k2 == k3) * (s1 == -s2) * ParticleOperator(
                            #     'b' + str(i1) + '^ d' + str(i4) + '^ a' + str(i2) + ' a' + str(i3)
                            # )
    
    return g**2 / (2 * L) * Y_four
```

```python
def Yukawa(I, m = 1, g = 1, L = 10):
    Y_int = ParticleOperator({})
    instFerm = ParticleOperator({})

    ham_Yukawa = H_free(I) + three_point(I, m, g, L) + four_point(I, m, g, L)

    return ham_Yukawa
    
```

```python
def generate_yukawa_circuit(number_of_modes, maximum_bosonic_occupation):

    operator =  Yukawa(number_of_modes)
    terms = operator.to_list()

    bosonically_rescaled_terms, bosonic_rescaling_factor = bosonically_rescale_terms(terms, maximum_bosonic_occupation)
    coefficients = [term.coeff for term in bosonically_rescaled_terms]


    norm = sum(np.abs(coefficients))
    target_state = get_target_state(coefficients)
    asp_rescaling_factor = (bosonic_rescaling_factor * norm)

    number_of_modes = max([term.max_mode() for term in terms]) + 1

    number_of_ancillae = (
        1000  # Some arbitrary large number with most ancilla disregarded
    )
    number_of_index_qubits = max(int(np.ceil(np.log2(len(terms)))), 1)
    number_of_rotation_qubits = max(get_number_of_active_bosonic_modes(terms)) + 1
    
    usp_rescaled_terms, usp_rescaling_factor = rescale_terms_usp(bosonically_rescaled_terms)
    usp_rescaling_factor *= (bosonic_rescaling_factor * (1 << number_of_index_qubits))

    # Declare Qubits
    validation = cirq.LineQubit(0)
    clean_ancillae = [cirq.LineQubit(i + 1) for i in range(number_of_ancillae)]
    rotation_qubits = [
        cirq.LineQubit(i + 1 + number_of_ancillae)
        for i in range(number_of_rotation_qubits)
    ]
    index_register = [
        cirq.LineQubit(i + 1 + number_of_ancillae + number_of_rotation_qubits)
        for i in range(number_of_index_qubits)
    ]
    system = System(
        number_of_modes=number_of_modes,
        maximum_occupation_number=maximum_bosonic_occupation,
        number_of_used_qubits=1
        + number_of_ancillae
        + number_of_rotation_qubits
        + number_of_index_qubits,
        has_fermions=operator.has_fermions,
        has_antifermions=operator.has_antifermions,
        has_bosons=operator.has_bosons,
    )
    
    #### USP Circuit Generation
    USP_numerics = {
        "left_elbows": 0,
        "right_elbows": 0,
        "rotations": 0,
        "ancillae_tracker": [1 + number_of_rotation_qubits + number_of_index_qubits],
        "angles": [],
        "number_of_nonclifford_rotations": 0,
        "rescaling_factor": usp_rescaling_factor,
        "number_of_modes": number_of_modes,
        "maximum_bosonic_occupation": maximum_bosonic_occupation,
        "maximum_number_of_active_bosonic_modes": number_of_rotation_qubits - 1,
        "number_of_terms": len(terms),
    }
    circuit = cirq.Circuit()
    circuit.append(cirq.I.on_each(*system.fermionic_register))
    circuit.append(cirq.I.on_each(*system.antifermionic_register))
    for bosonic_reg in system.bosonic_system:
        circuit.append(cirq.I.on_each(*bosonic_reg))
    circuit.append(cirq.X.on(validation))
    circuit += add_naive_usp(index_register)
    circuit += add_lobe_oracle(
        usp_rescaled_terms,
        validation,
        index_register,
        system,
        rotation_qubits,
        clean_ancillae,
        perform_coefficient_oracle=True,
        decompose=True,
        numerics=USP_numerics,
    )
    circuit += add_naive_usp(index_register)
    USP_numerics["number_of_ancillae"] = max(USP_numerics["ancillae_tracker"])
    USP_numerics["number_of_qubits"] = max(USP_numerics["ancillae_tracker"]) + system.number_of_system_qubits
    for angle in USP_numerics["angles"]:
        angle = np.abs(angle)
        if not np.isclose(angle%np.pi/4, 0):
            USP_numerics["number_of_nonclifford_rotations"] += 1
    
    
    #### ASP Circuit Generation
    ASP_numerics = {
        "left_elbows": 0,
        "right_elbows": 0,
        "rotations": 0,
        "ancillae_tracker": [1 + number_of_rotation_qubits - 1 + number_of_index_qubits],
        "angles": [],
        "number_of_nonclifford_rotations": 0,
        "rescaling_factor": asp_rescaling_factor,
        "number_of_modes": number_of_modes,
        "maximum_bosonic_occupation": maximum_bosonic_occupation,
        "maximum_number_of_active_bosonic_modes": number_of_rotation_qubits - 1,
        "number_of_terms": len(terms)
    }
    circuit = cirq.Circuit()
    circuit.append(cirq.I.on_each(*system.fermionic_register))
    circuit.append(cirq.I.on_each(*system.antifermionic_register))
    for bosonic_reg in system.bosonic_system:
        circuit.append(cirq.I.on_each(*bosonic_reg))
    circuit.append(cirq.X.on(validation))
    circuit += add_prepare_circuit(
        index_register,
        target_state=target_state,
        numerics=ASP_numerics,
        clean_ancillae=clean_ancillae
    )
    circuit += add_lobe_oracle(
        bosonically_rescaled_terms,
        validation,
        index_register,
        system,
        rotation_qubits,
        clean_ancillae,
        perform_coefficient_oracle=False,
        decompose=True,
        numerics=ASP_numerics,
    )
    circuit += add_prepare_circuit(
        index_register,
        target_state=target_state,
        dagger=True,
        numerics=ASP_numerics,
        clean_ancillae=clean_ancillae
    )
    ASP_numerics["number_of_ancillae"] = max(ASP_numerics["ancillae_tracker"])
    ASP_numerics["number_of_qubits"] = max(ASP_numerics["ancillae_tracker"]) + system.number_of_system_qubits
    for angle in ASP_numerics["angles"]:
        angle = np.abs(angle)
        if not np.isclose(angle%np.pi/4, 0):
            ASP_numerics["number_of_nonclifford_rotations"] += 1
    
    return {
        "USP": USP_numerics,
        "ASP": ASP_numerics 
    }
```
